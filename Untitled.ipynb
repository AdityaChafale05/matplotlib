```python
import matplotlib.pyplot as plt 


months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
#prices of Ram in 2024-25
prices_2025 = [3200,3300,3450,3600,3800,3950,4100,4050,3900,3700,3550,3400]
prices_2024 = [1100,1120,1180,1240,1290,1250,1250,350,420,560,700,850]

plt.plot(months,prices_2025,color="red",linestyle=':',marker='o',label='2025')
plt.plot(months,prices_2024,color="blue",linestyle='--',marker='^',label='2024')

plt.xlabel("Months")
plt.ylabel("Cost")
plt.title("RAM prices 2024-25 (ddr4 8GB)")
plt.grid(True)
plt.legend();

!git init
```
