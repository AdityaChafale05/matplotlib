```python
import matplotlib.pyplot as plt 


years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.xlabel(years)
plt.ylabel(runs)
plt.title("performance")
#plt.barh(years,runs,color="green");
plt.bar(years,runs,label="runs",color="green")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli  = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

sel=[1999,2001,2005,2006,2009,2010]

n = len(sel)
sachin_plot = sachin[:n]
sehwag_plot = sehwag[:n]
kohli_plot = kohli[:n]

x = np.arange(n)  # index positions
width = 0.25

plt.bar(x-width,sachin_plot, width=width, color="yellow", label="sachin")
plt.bar(x,sehwag_plot, width=width, color="skyblue", label="sehwag")
plt.bar(x+width,kohli_plot, width=width, color="#8a1926", label="kohli")

plt.xticks(x,sel)


for i in range(n):
   plt.text(i,sachin[i]+50,str(sachin[i]), ha="right",color="#ccc92d")
   

for i in range(n):
   plt.text(i,sehwag[i]+50,str(sehwag[i]), ha="left",color="blue")
 
for i in range(n):
   plt.text(i,kohli[i]+50,str(kohli[i]), ha="center",color="red")
 

plt.xlabel("players")
plt.ylabel("runs")
plt.title("Comparison")
plt.tight_layout()
plt.legend();
```
