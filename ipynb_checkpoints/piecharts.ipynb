```python
import matplotlib.pyplot as plt 

plt.style.use("ggplot") 

# Data
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
explode=[0.2,0,0,0]
colors=['#F20F0F','#1F0F0D','#FFC50A','#C2A3B3']
plt.title("Career Runs of Indian Batsmen")
plt.pie(runs, labels=labels, colors=colors, explode = explode, autopct = '%1.1f%%',wedgeprops = { 'edgecolor' : 'red', 'linestyle': '--','linewidth': 1},shadow=True)
plt.show()
```
